```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('general_data.csv')
df.head()
```

```python
df.isna().sum()
```

```python
df = df.fillna(method='ffill')
```

```python
df.isna().sum()
```

```python
df = df.drop(columns = ['EmployeeID', 'EmployeeCount', 'Over18', 'StandardHours', ])
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
def encode(cols):
    lc = LabelEncoder()
    for col in cols:
        df[col] = lc.fit_transform(df[col])
ecol = ['Attrition', 'BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus']
encode(ecol)
```

```python
x = df.drop(columns = ['Attrition'])
y = df.iloc[:,1]
```

```python
import statsmodels.api as sm
```

```python
co = sm.add_constant(x)
logi = sm.Logit(y, co)
```

```python
pred = logi.fit()
```

```python
pred.summary()
```
