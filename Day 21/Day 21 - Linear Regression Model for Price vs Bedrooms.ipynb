```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_excel("Linear Regression.xlsx")
```

```python
x = df1.bedrooms
y = df1.price
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
X_train= X_train.values.reshape(-1, 1)
y_train= y_train.values.reshape(-1, 1)
X_test = X_test.values.reshape(-1, 1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
l_reg = LinearRegression()
```

```python
lreg = l_reg.fit(X_train,y_train)
```

```python
plt.scatter(X_train,y_train)
plt.plot(X_train,l_reg.predict(X_train),color = 'black')
```

```python
ypred = l_reg.predict(X_test)
```

```python
plt.scatter(X_test,y_test)
plt.plot(X_test,ypred,color = 'black')
```

```python
from sklearn.metrics import r2_score
print("R2 score of the Price vs Bedrooms Model is - ",r2_score(y_test,ypred))
```
